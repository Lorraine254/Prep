```python
#Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Fetching the data
cons_data=pd.read_csv('https://raw.githubusercontent.com/chrisnjunge/DStest/master/GuaraldiConsultants.xlsx%20-%20Sheet2.csv')
cons_data.head()
```

a. Create a scatter chart and add a trendline to examine the relationship
between years of service and amount of vacation time used for all
consultants at Guaraldi and Associates. Based on the scatter charts, what
appears to be the relationship between years of service and amount of
vacation time used? (5 marks)

```python
# Defining the plotting style
plt.style.use('seaborn')

# Extract data from the DataFrame
years_of_service = cons_data['Years of Service']
vacation_time_used = cons_data['Vacation Hours Used']

plt.scatter(years_of_service,vacation_time_used,alpha=0.75,c='green')

# Calculate the trendline
m, b = np.polyfit(years_of_service, vacation_time_used, 1)

# Add the trendline to the plot
plt.plot(years_of_service, m * years_of_service + b)


# Add labels and title
plt.xlabel("Years of Service")
plt.ylabel("Vacation Time Used")
plt.title('Scatter Plot of Years of Service and Vacation Hours')
plt.show
```

**Insight**
- There seems to be a negative relationship between years of service and vacation time used.

### b. Create a scatter chart for the same data, but this time differentiate the marks in

**Insight**
- The junior consultants seem to spend less time on vacation

- Data Understanding - the data available used to examine this is years of Service,	Vacation Hours Used	,Consultant Title. Vacation hours used is the number of hours used outside the office, while the consultant title is the position of the consultant in the company, which could be a senior or junior position.

- Data preparation- this entails cleaning the data by dealing with missing values and outliers which may affect statistical analysis. An example is a dataset that contains outliers, the best measure of central tendency to use in order to summarise the data is the median since it is robust to outliers.

- Data modelling- this refers to identifying what modeling technique we can use in our data set. A classification model can be used to classify the position of the consultants as junior or senior.
